파이썬에서 함수 내부의 대입문은 지역 변수를 생성하지만,
가변 객체의 내부 상태를 변경하는 연산은
동일 객체에 대한 참조를 통해 전역에도 영향을 미친다.

```python
a = 3

def func():
    print(a)

func()
```

```python
b = 4
def func2():
    b = 5 # 밖의 b = 4를 변경한게 아니라, 함수 공간에 b를 새로 만든 것
    print(b)
func2()
print(b)
```

`lst[0] = 100`은 전역 변수가 가리키는 객체의 내부 값을 수정하는 것

```python
lst = [1, 2, 3]

def func3():
    lst[0] = 100
    print(lst)

func3()
print(lst)
```

반면 `lst = [...]`는 지역 변수에 새 객체를 재할당하는 것이다

```python
lst = [1, 2, 3]

def func3():
    lst = [100, 2, 3]
    print(lst)

func3()
print(lst)
```
